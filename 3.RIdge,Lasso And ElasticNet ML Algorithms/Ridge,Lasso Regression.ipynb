```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv',header = 1)
```

```python
dataset.head()
```

```python
#First check their datatypes
dataset.info()
```

### Data Cleaning

```python
## missing values
dataset.isnull().sum() #or we can do
dataset[dataset.isnull().any(axis = 1)]
```

The dataset is converted in to two sets based on region 122th index , we can  make a new column based on the Region (mentioned in the csv file)  

```python
dataset.loc[:122,"Region"] = 0
dataset.loc[122:,"Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
## Removing the null values
df = df.dropna().reset_index(drop = True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
##Removing the 122nd row
df = df.drop(122).reset_index(drop = True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
##Fixing spaces in column name
df.columns = df.columns.str.strip()
```

```python
df.columns
```

### Changes the required columns as  interger data type

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws',]] = df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws',]].astype(int)
```

```python
df.info()
```

### Changes the other columns as  float data type

```python
objects = [ I for I in df.columns if df[I].dtypes == 'O']
```

```python
for i in objects:
    if i!= 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.head()
```

```python
## Let save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index= False)
```
